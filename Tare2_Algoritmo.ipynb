```python
%%writefile R1.cpp
#include <iostream>
using namespace std;
void Lectura(int n, int arreglo[]) {//BigO(n)
    for (int i = 0; i < n; i++) {//1+n(1+6+2)->1+n(9) BigO(n)
        cout << "NUMERO " << i + 1 << ": ";//1+1+1+1=4
        cin >> arreglo[i];//2
    }
}
void Mayor(int n, int arreglo[]) {//BigO(n)
    int mayor = arreglo[0];//2
    for (int i = 0; i < n; i++) {//1+n(1+4+2)
        if (arreglo[i] > mayor) {//2
            mayor = arreglo[i];//2
        }
    }
    cout << "EL NUMERO MAYOR ES: " << mayor << endl;//3
}//2+1+n(1+4+2)+3->6+n(7) BigO(n)
void Impresion(int n, int arreglo[]) {//BigO(n)
    for (int i = 0; i < n; i++) {//1+n(1+3+2)
        cout << arreglo[i] << " ";//3
    }
    cout << endl;//1
}//1+n(6)+1->2+n(6) BigO(n)
void Ordenamiento(int n, int arreglo[]) {//BigO(n2)
    int aux = 0;//1
    for (int i = 0; i < n; i++) {//1+n(1+2+n(13)+2)+1->2+n(5+13(n))->2+5n+13n2 BigO(n2)
        for (int j = i + 1; j < n; j++) {//2+n(1+10+2)->2+n(13)
            if (arreglo[i] < arreglo[j]) {//3
                aux = arreglo[i];//2
                arreglo[i] = arreglo[j];//3
                arreglo[j] = aux;//2
            }
        }
    }
    Impresion(n, arreglo);//BigO(n)
    for (int i = 0; i < n; i++) {//1+n(1+1+n(14)+2)->1+n(4+14(n))->1+4n+14(n2) BigO(n2)
        for (int j = i + 1; j < n; j++) {//1+n(2+10+2)->1+n(14)
            if (arreglo[i] > arreglo[j]) {//3
                aux = arreglo[i];//2
                arreglo[i] = arreglo[j];//3
                arreglo[j] = aux;//2
            }
        }
    }
    Impresion(n, arreglo);//BigO(n)
}
void Eliminacion(int n, int arreglo[]) {//7+1+4n+6(n2)+2->10+4n+6(n2) BigO(n2)
    int posicion = 0;//1
    cout << "POSICION A ELIMINAR: ";//1
    cin >> posicion;//1
    cout << "NUMERO A ELIMINAR: " << arreglo[posicion] << endl;//4
    for (int i = 0; i < n; i++) {//1+n(1+1+n(6)+2)->1+n(4+n(6))->1+4n+6(n2)
        if (i == posicion) {//1
            while (posicion < n) {//n(4+2)->n(6)
                arreglo[posicion] = arreglo[posicion + 1];//4
                posicion++;//2
            }
        }
    }
    n = n - 1;//2
    Impresion(n, arreglo);//BigO(n)
}
void Busqueda(int n, int arreglo[]) {//4+1+n(6)+9->14+6(n) BigO(n)
    bool encontrado = false;//1
    int numero = 0;//1
    cout << "NUMERO A BUSCAR: ";//1
    cin >> numero;//1
    for (int i = 0; i < n; i++) {//1+n(1+3+2)->1+n(6)
        if (arreglo[i] == numero) {//2
            encontrado = true;//1
        }
    }
    if (encontrado == true) {//1
        cout << "EL NUMERO " << numero << " SE ENCOTRO " << endl;//4
    }
    else {
        cout << "EL NUMERO " << numero << " NO SE ENCONTRO" << endl;//4
    }
}
void Factorial() {//4+1+n(5)+5->10+5(n) BigO(n)
    int numero = 0;//1
    cout << "NUMERO: ";//1
    cin >> numero;//1
    int factorial = 1;//1
    for (int i = 1; i <= numero; i++) {//1+n(1+2+2)->1+n(5)
        factorial = factorial * i;//2
    }
    cout << "EL FACTORIAL DE " << numero << " ES: " << factorial << endl;//5
}
void Contador(int n, int arreglo[]) {//4+1+n(7)+6->11+n(7) BigO(n)
    int cont = 0;//1
    int numero = 0;//1
    cout << "NUMERO A VER: ";//1
    cin >> numero;//1
    for (int i = 0; i < n; i++) {//1+n(1+4+2)->1+n(7)
        if (numero == arreglo[i]) {//2
            cont++;//2
        }
    }
    cout << "EL NUMERO " << numero << " SE REPITE " << cont << " VECES" << endl;//6
}
void Suma() {//4+n(6)+3->7+6(n) BigO(n)
    int numero = 0;//1
    cout << "NUMERO: ";//1
    cin >> numero;//1
    int suma = 0;//1
    while (numero > 0) {//n(1+5)
        suma = suma + numero % 10;//3
        numero = numero / 10;//2
    }
    cout << "LA SUMA ES: " << suma << endl;//3
}
void EsPrimo() {//4+1+n(7)+5->10+7(n) BigO(n)
    int numero = 0;//1
    cout << "NUMERO: ";//1
    cin >> numero;//1
    int cont = 0;//1
    for (int i = 1; i <= numero; i++) {//1+n(1+4+2)->1+n(7)
        if (numero % i == 0) {//2
            cont++;//2
        }
    }
    if (cont == 2) {//1
        cout << "NUMERO PRIMO" << endl;//2
    }
    else {
        cout << "NUMERO NO PRIMO" << endl;//2
    }
}
void Primo(int n, int arreglo[]) {//1+1+n(11+log(n)(8))+4->6+n(11)+n(log(n)(8)) BigO(n(log(n)))
    int aux = 0;//1
    for (int i = 0; i < n; i++) {//1+n(1+1+1+log(n)(8)+6+2)->1+n(11+log(n)(8))
        int cont = 0;//1
        for (int j = 1; j <= arreglo[i]; j++) {//1+log(n)(1+5+2)->1+log(n)(8)
            if (arreglo[i] % j == 0) {//3
                cont++;//2
            }
        }
        if (cont == 2) {//1
            cout << arreglo[i] << " ";//3
            aux++;//2
        }
    }
    cout << endl;//1
    cout << "LA CANTIDAD DE NUMERO PRIMOS SON: " << aux << endl;//3
}
int main() {//37+9*BigO(n)+BigO(n(log(n)))+2*BigO(n^2)=BigO(n^2)
    int n;//1
    int arreglo[100];//1
    cout << "CANTIDAD DE TERMINOS: ";//1
    cin >> n;//1
    cout << endl;//1
    cout << "PREGUNTA 1 (NUMERO MAYOR): " << endl;//2
    Lectura(n, arreglo);//BigO(n)
    cout << endl;//1
    Impresion(n, arreglo);//BigO(n)
    cout << endl;//1
    Mayor(n, arreglo);//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 2 (ORDENAMIENTO): " << endl;//2
    Ordenamiento(n, arreglo);//BigO(n^2)
    cout << endl;//1
    cout << "PREGUNTA 3 (ELIMINAR NUMERO DE UNA POSICION): " << endl;//2
    Eliminacion(n, arreglo);//BigO(n^2)
    cout << endl;//1
    cout << "PREGUNTA 4 (BUSCAR UN NUMERO): " << endl;//2
    Busqueda(n, arreglo);//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 5 (FACTORIAL DE UN NUMERO): " << endl;//2
    Factorial();//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 6 (VER SI UN NUMERO EXISTE O NO): " << endl;//2
    Busqueda(n, arreglo);//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 7 (CALCULO DE VECES DE UN NUMERO): " << endl;//2
    Contador(n, arreglo);//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 8 (SUMA DE DIGITOS DE UN NUMERO): " << endl;//2
    Suma();//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 9 (DETERMINAR SI UN NUMERO ES PRIMO O NO): " << endl;//2
    EsPrimo();//BigO(n)
    cout << endl;//1
    cout << "PREGUNTA 10 (CANTIDAD DE NUMEROS PRIMOS): " << endl;//2
    Primo(n, arreglo);//BigO(n(log(n)))
    cout << endl;//1
}
```
