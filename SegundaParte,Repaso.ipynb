```python
%%writefile R1.cpp
#include <iostream>
using namespace std;
void SumaPromedio(int n, int A[]) {//2+1+n(9)+3->6+n(9) BigO(n)
    int s = 0;//1
    double prom = 0;//1
    for (int i = 0; i < n; i++) {//1+n(1+6+2)->1+n(9)
        s = s + A[i];//3
        prom = s / (double)n;//3
    }
    printf("Suma:%d\nProm:%d", s, prom);//3
}
void Ejercicio2(int n, int A[]) {//2+1+n(6)+6->9+n(6) BigO(n)
    int s = 0;//1
    double prom = 0;//1
    for (int i = 0; i < n; i++) {//1+n(1+3+2)->1+n(6)
        s = s + A[i];//3
    }
    prom = s / (double)n;//3
    printf("Suma:%d\nProm:%d", s, prom);//3
}
void Ordenamiento(int n,int vec[]) {//1+4n+13(n)(n-1)->1+4n+13n^2-13(n) BigO(n^2)
    for (int i = 0; i < n - 1; i++) {//1+n(1+1+(n-1)(13)+2)->1+n(4+(n-1)(13))
        for (int k = i + 1; k < n; k++) {//1+(n-1)(1+10+2)->1+(n-1)(13)
            if (vec[i] > vec[k]) {//3
                int aux = vec[i];//2
                vec[i] = vec[k];//3
                vec[k] = aux;//2
            }
        }
    }
}
void Ordenamiento2(int n, int vec[]) {//2+n(9)+BigO(1) BigO(n)
    int frec[101] = { 0 };//1
    for (int i = 0; i < n; i++) {//1+n(1+6+2)->1+n(9)+1->2+n(9)
        frec[vec[i]]++;//frec[vec[i]]=frec[vec[i]]+1//6
    }
    int pos = 0;//1
    for (int i = 0; i < 101; i++) {//1+101(1+1+log(101)(4)+2)->1+101(4+log(101)(4))+1->2+101(4+log(101)(4))
        for (int k = 0; k < frec[i]; k++) {//1+log(101)(4)
            vec[pos] = i;//2
            pos++;//2
        }
    }
}
void Algoritmo(int n, int vec[]) {//2+100(5+5(n)+2log(n))=BigO(n)
    int i = 0;//1
    int sum = 0;//1
    while (i < 100) {//100(1+2+1+(5)n+1+(2(log)(n)))->100(5+5(n)+2log(n))
        if (i % 2 == 0) {//2
            for (int k = 0; k < n; k++) {//1+n(1+2+2)->1+n(5)
                sum += vec[i];//2
            }
        }
        else {
            for (int k = 0; k < i; k++) {//1+log(n)(2)
                sum += vec[i];//2
            }
        }
    }
}
void Factorial(int n) {//1+1+5n-10+1->5n-7 BigO(n)
    int fact = 1;//1
    for (int i = 2; i < n; i++) {//1+(n-2)(1+2+2)->1+(n-2)(5)
        fact *= i;//2
    }
    printf("Factorial: %d", fact);//1
}
void Busqueda(int n, char* vec[], char *cadBuscar) {//1+1+(8)n->2+8n BigO(n)
    int pos = -1;//1
    for (int i = 0; i < n; i++) {//1+n(1+5+2)->1+n(8)
        if (strcmp(vec[i], cadBuscar) == 0) {//3
            pos = i;//1
            break;//1
        }
    }
}
void Mayor(int n, int* vec[]) {//1+7n-6+2->7n-3 BigO(n)
    int pos = 0;//1
    for (int i = 1; i < n; i++) {//1+(n-1)(1+4+2)->1+7n-7->7n-6
        if (vec[i] > vec[pos])//3
            pos = i;//1
    }
    printf("El mayor es: %d", vec[pos]);//2
}
void AlgoritmoRaro(int n, int* vec[]) {//1+1+n(16)+2->4+n(16) BigO(n)
    int max = 0;//1
    for (int m = 0; m < n; m++) {//1+n(1+13+2)->1+n(16)
        int cont = 0;//1
        int k = m + 1;//2
        while (vec[m] <= vec[k]) {//3
            k = k + 1;//2
            cont++;//2
        }
        if (cont > max)//1
            max = cont;//2
    }
    printf("Maximo %d", max);//2
}
void Logaritmica(int n, int vec[]) {//1+8log(n) BigO(log(n))
    int i = 1;//1
    while (i < n) {//log(n)(1+7)->8log(n)
        if (vec[i] % 2 == 0)//3
            i *= 3;//2
        else
            i *= 2;//2
    }
}
void Logaritmica2(int n, int vec[]) {//1+8log(n) BigO(log(n))
    int i = 1;//1
    while (i < n) {//log(n)(1+7)->8log(n)
        if (vec[i] % 2 == 0)//3
            n = n / 3;//2
        else
            n = n / 2;//2
    }
}
void BusquedaBinaria(int n, int arreglo[]) {//4+log(n+15) BigO(log(n))
    int inf = 0;//1
    int sup = n - 1;//2
    int pos = -1;//1
    while ((pos == -1) && (sup >= inf)) {//log(n+3+12)->log(n+15)
        int medio = (inf + sup) / 2;//3
        if (arreglo[medio] == 80)//2
            pos = medio;//1
        else if (arreglo[medio] < 80)//2
            inf = medio + 1;//2
        else
            sup = medio - 1;//2
    }
}
int main()
{
    
}
```
